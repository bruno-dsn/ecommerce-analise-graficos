```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual
sns.set(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (10, 6)

# 1) Ler o arquivo (use exatamente o nome do arquivo que você anexou/tem no Colab)
df = pd.read_csv('ecommerce_estatistica.csv')

# 2) Checagem básica
print("Dimensão do dataset:", df.shape)
print("\nColunas:")
print(list(df.columns))

print("\nTipos de dados:")
print(df.dtypes)

print("\nPrévia dos dados:")
df.head()
```

```python
# 3) Padronizar tipos numéricos (garante que colunas numéricas estejam como número)
num_candidates = [
    'Nota', 'N_Avaliações', 'Desconto', 'Preço',
    'Nota_MinMax', 'N_Avaliações_MinMax', 'Desconto_MinMax', 'Preço_MinMax',
    'Marca_Cod', 'Material_Cod', 'Temporada_Cod',
    'Qtd_Vendidos', 'Qtd_Vendidos_Cod', 'Marca_Freq', 'Material_Freq'
]

for col in num_candidates:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# 4) Resumo estatístico das variáveis numéricas
df.describe(include=[np.number]).T
```

```python
# Histograma da distribuição de preço (usa 'Preço' se existir; senão, 'Preço_MinMax' como fallback)
col_preco = 'Preço' if 'Preço' in df.columns else ('Preco' if 'Preco' in df.columns else 'Preço_MinMax')

plt.figure()
sns.histplot(df[col_preco].dropna(), kde=False, bins=30, color='steelblue')
plt.title('Distribuição de preço')
plt.xlabel('Preço')
plt.ylabel('Contagem')
plt.tight_layout()
plt.show()
```

```python
# Dispersão: preço vs número de avaliações (usa 'N_Avaliações' como padrão)
col_av = 'N_Avaliações' if 'N_Avaliações' in df.columns else ('N_Avaliacoes' if 'N_Avaliacoes' in df.columns else 'N_Avaliações_MinMax')

plt.figure()
sns.scatterplot(x=df[col_preco], y=df[col_av], alpha=0.6)
plt.title('Dispersão: preço vs número de avaliações')
plt.xlabel('Preço')
plt.ylabel('Número de avaliações')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de calor das correlações (variáveis numéricas)')
plt.tight_layout()
plt.show()
```

```python
if 'Marca' in df.columns:
    marca_freq = df['Marca'].value_counts().head(15)
    plt.figure()
    sns.barplot(x=marca_freq.values, y=marca_freq.index, orient='h', color='teal')
    plt.title('Frequência de marcas (Top 15)')
    plt.xlabel('Contagem')
    plt.ylabel('Marca')
    plt.tight_layout()
    plt.show()
else:
    print("Coluna 'Marca' não encontrada. Ajuste o nome se necessário.")
```

```python
if 'Material' in df.columns:
    material_freq = df['Material'].value_counts().head(8)
    plt.figure()
    plt.pie(material_freq.values, labels=material_freq.index, autopct='%1.1f%%', startangle=90)
    plt.title('Distribuição de materiais (Top 8)')
    plt.tight_layout()
    plt.show()
else:
    print("Coluna 'Material' não encontrada. Ajuste o nome se necessário.")
```

```python
# Densidade do campo de desconto (usa 'Desconto'; caso contrário, 'Desconto_MinMax')
col_desc = 'Desconto' if 'Desconto' in df.columns else 'Desconto_MinMax'

plt.figure()
sns.kdeplot(df[col_desc].dropna(), fill=True, color='purple', alpha=0.3)
plt.title('Densidade do desconto')
plt.xlabel('Desconto')
plt.ylabel('Densidade')
plt.tight_layout()
plt.show()
```

```python
# Regressão entre preço e quantidade vendida ('Qtd_Vendidos_Cod' geralmente é melhor para regressão simples)
col_qtd = 'Qtd_Vendidos_Cod' if 'Qtd_Vendidos_Cod' in df.columns else ('Qtd_Vendidos' if 'Qtd_Vendidos' in df.columns else None)

if col_qtd is not None:
    plt.figure()
    sns.regplot(x=df[col_preco], y=df[col_qtd], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
    plt.title('Regressão: preço vs quantidade vendida')
    plt.xlabel('Preço')
    plt.ylabel('Quantidade vendida (codificada)' if col_qtd=='Qtd_Vendidos_Cod' else 'Quantidade vendida')
    plt.tight_layout()
    plt.show()
else:
    print("Nenhuma coluna de quantidade encontrada ('Qtd_Vendidos_Cod' ou 'Qtd_Vendidos').")
```
